## Visualization

```python
corr_df=df_spotify.corr(method="pearson")
plt.figure(figsize=(14,6))
heatmap=sns.heatmap(corr_df, annot=True, cmap="icefire")
```

```python
plt.figure(figsize=(8,4))
sns.regplot(data = df_spotify, y = "danceability", x = "popularity", color = "c").set(title = "Popularity vs Danceability")
```

```python
plt.figure(figsize=(8,4))
sns.regplot(data = df_spotify, y = "loudness", x = "popularity", color = "c").set(title = "Popularity vs Loudness")
```

```python
plt.figure(figsize=(8,4))
sns.regplot(data = df_spotify, y = "acousticness", x = "popularity", color = "orange").set(title = "Popularity vs Acousticness")
```

```python
# To check for the top genre by its popularity
plt.figure(figsize=(12,8))
sns.barplot(data = df_spotify,y = "genre", x = "popularity", color='r').set(title="Graph showing top Genre by popularity")
```

```python
sns.scatterplot(x="popularity",y="danceability",data = df_spotify,hue="energy")
```

```python
plt.figure(figsize=(18,4))
sns.histplot(df_spotify["genre"], kde=True, bins=10 )
```
